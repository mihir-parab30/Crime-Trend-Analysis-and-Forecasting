```python
# Handeling missing value with mode method
df['Mocodes'] = df['Mocodes'].fillna(df['Mocodes'].mode().iloc[0])
df['Vict Sex'] = df['Vict Sex'].fillna(df['Vict Sex'].mode().iloc[0])
df['Vict Descent'] = df['Vict Descent'].fillna(df['Vict Descent'].mode().iloc[0])
df['Weapon Desc'] = df['Weapon Desc'].fillna(df['Weapon Desc'].mode().iloc[0])
df['Status']=df['Status'].fillna(df['Status'].mode().iloc[0])
df['Cross Street']=df['Cross Street'].fillna(df['Cross Street'].mode().iloc[0])
df['Premis Desc'] = df['Premis Desc'].fillna(df['Premis Desc'].mode().iloc[0])
```

```python
# Handeling missing value with median method
df['Premis Cd']=df['Premis Cd'].fillna(df['Premis Cd'].median())
df['Weapon Used Cd']=df['Weapon Used Cd'].fillna(df['Weapon Used Cd'].median())
df['Crm Cd 1']=df['Crm Cd 1'].fillna(df['Crm Cd 1'].median())
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Dropping columns with high percentage of null values
df.drop(columns=['Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4'],inplace=True)
```

```python
# dropping duplicate rows
df.drop_duplicates().head(10)
```

```python
# converting object data type to date & time datatypes
from datetime import datetime as datetime

df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])

time_occ_list = df['TIME OCC'].apply(lambda x: str(x).zfill(4))
time_list = []

# convert 'TimeInt' column to time format
for time_item in time_occ_list:
     time_list.append( (datetime.strptime(str(time_item),"%H%M").time()).strftime("%H:%M"))

df['TIME OCC'] = time_list

df[['Date Rptd', 'DATE OCC' , 'TIME OCC']]
```

```python
df.set_index('DATE OCC',inplace=True)
df.sort_index(inplace=True)
crime=df.resample('M').size()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Fit an ARIMA(1,1,1) model to the 'Crime Count' time series
model = ARIMA(crime, order=(1, 1, 1))
results = model.fit()

# Print the model summary
print(results.summary())
```

```python
# forecast the next 12 months 
forecast_steps=12
forecast=results.get_forecast(steps=forecast_steps)
# get forecasted values and their intervals
forecast_values=forecast.predicted_mean
confidence_intervals=forecast.conf_int()
# print the forecasted values and their intervals
print('Forecasted Crime Counts:')
print(forecast_values)
print('\nConfidence Intervals:')
print(confidence_intervals)
```

```python
# plot the actual observed data
plt.figure(figsize=(10,6))
plt.plot(crime.index,crime.values,label='Actual',marker='o')
#plot the forecasted values
plt.plot(forecast_values.index,forecast_values.values,color='red',label='Forecast',marker='o')
# customize the plot
plt.fill_between(confidence_intervals.index,confidence_intervals.iloc[:,0],confidence_intervals.iloc[:,1],color='pink',alpha=0.3,label='confidence Intervals')
plt.title('Crime Forcasting')
plt.xlabel('Date')
plt.ylabel('Crime Counts')
plt.legend()
plt.grid(True)
plt.show()
```
